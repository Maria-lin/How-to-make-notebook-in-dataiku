## 5. ANOMALIE 1: Montant moyen anormal

```python
if 'type_carte' in df.columns and 'montant_moyen_retrait' in df.columns:
    print("=== ANOMALIE 1: MONTANT MOYEN ANORMAL ===")
    print("")
    print("Automates avec montant moyen tres different de la mediane")
    print("pour un type de carte (comportement incoherent)")
    print("")

    anomalies_list = []

    # Pour chaque type de carte
    for type_c in df['type_carte'].unique():
        subset = df[df['type_carte'] == type_c].copy()

        # Stats
        median_val = subset['montant_moyen_retrait'].median()
        std_val = subset['montant_moyen_retrait'].std()

        # Seuils: +/- 2 ecarts-types
        seuil_bas = median_val - 2.0 * std_val
        seuil_haut = median_val + 2.0 * std_val

        # Detecter
        mask = (subset['montant_moyen_retrait'] < seuil_bas) | (subset['montant_moyen_retrait'] > seuil_haut)
        anomalies = subset[mask].copy()

        if len(anomalies) > 0:
            anomalies['type_analyse'] = type_c
            anomalies['median_ref'] = median_val
            anomalies['ecart'] = anomalies['montant_moyen_retrait'] - median_val
            anomalies_list.append(anomalies)

    if len(anomalies_list) > 0:
        df_anom1 = pd.concat(anomalies_list, ignore_index=True)
        print("Anomalies detectees: " + str(len(df_anom1)))
        print("")
        print("Top 20:")

        # Trier par ecart absolu
        df_anom1['ecart_abs'] = df_anom1['ecart'].abs()
        df_anom1_sorted = df_anom1.sort_values('ecart_abs', ascending=False)

        cols = ['num_automate', 'type_analyse', 'montant_moyen_retrait', 'median_ref', 'ecart']
        cols = [c for c in cols if c in df_anom1_sorted.columns]
        print(df_anom1_sorted[cols].head(20))
    else:
        print("Aucune anomalie")
```

## 6. ANOMALIE 2: Desequilibre extreme

```python
if 'num_automate' in df.columns and 'type_carte' in df.columns:
    print("=== ANOMALIE 2: DESEQUILIBRE EXTREME ===")
    print("")
    print("Automates ou un type domine excessivement (>90%)")
    print("")

    concentration_list = []

    # Pour chaque automate
    for automate_id in df['num_automate'].unique():
        subset = df[df['num_automate'] == automate_id]
        total_ret = subset['nb_total_de_retraits'].sum()

        if total_ret > 0:
            for type_c in subset['type_carte'].unique():
                nb_type = subset[subset['type_carte'] == type_c]['nb_total_de_retraits'].sum()
                pct = (float(nb_type) / float(total_ret)) * 100.0

                concentration_list.append({
                    'num_automate': automate_id,
                    'type_carte': type_c,
                    'pct': pct,
                    'nb_type': nb_type,
                    'total': total_ret
                })

    df_conc = pd.DataFrame(concentration_list)

    # Max par automate
    max_conc = df_conc.groupby('num_automate')['pct'].max().reset_index()
    max_conc.columns = ['num_automate', 'max_pct']

    # Desequilibre > 90%
    deseq = max_conc[max_conc['max_pct'] > 90.0]
    print("Automates avec concentration >90%: " + str(len(deseq)))

    if len(deseq) > 0:
        # Detail
        deseq_detail = deseq.merge(df_conc[df_conc['pct'] > 90.0], on='num_automate')
        print("")
        print("Detail:")
        cols = ['num_automate', 'type_carte', 'pct', 'nb_type', 'total']
        print(deseq_detail[cols].head(20))

    # Visualisation
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111)
    max_conc['max_pct'].hist(bins=30, edgecolor='black', ax=ax)
    ax.set_xlabel('% max concentration')
    ax.set_ylabel('Nb automates')
    ax.set_title('Distribution concentration')
    ax.axvline(90.0, color='red', linestyle='--', label='Seuil 90%')
    ax.legend()
    plt.tight_layout()
    plt.show()
```

## 7. ANOMALIE 3: Ratio incoherent

```python
if 'num_automate' in df.columns and 'type_carte' in df.columns:
    print("=== ANOMALIE 3: RATIO INCOHERENT ===")
    print("")
    print("Automates ou le ratio montant/volume varie")
    print("de facon extreme entre types (>3x)")
    print("")

    # Automates multi-types
    multi_types = df.groupby('num_automate')['type_carte'].nunique()
    multi_types = multi_types[multi_types > 1].index

    ratio_anom = []

    for automate_id in multi_types:
        subset = df[df['num_automate'] == automate_id].copy()

        # Ratios par type
        ratios = {}
        for idx, row in subset.iterrows():
            if row['nb_total_de_retraits'] > 0:
                ratio = float(row['montant_total']) / float(row['nb_total_de_retraits'])
                ratios[row['type_carte']] = ratio

        if len(ratios) >= 2:
            ratio_vals = list(ratios.values())
            ratio_min = min(ratio_vals)
            ratio_max = max(ratio_vals)

            if ratio_min > 0:
                variation = ratio_max / ratio_min

                if variation > 3.0:
                    ratio_anom.append({
                        'num_automate': automate_id,
                        'ratio_min': ratio_min,
                        'ratio_max': ratio_max,
                        'variation': variation,
                        'nb_types': len(ratios)
                    })

    if len(ratio_anom) > 0:
        df_ratio = pd.DataFrame(ratio_anom)
        df_ratio = df_ratio.sort_values('variation', ascending=False)

        print("Automates avec variation >3x: " + str(len(df_ratio)))
        print("")
        print("Top 20:")
        print(df_ratio.head(20))
    else:
        print("Aucune anomalie")
```

```python
# Dashboard complet
if 'montant_total' in df.columns and 'nb_total_de_retraits' in df.columns:
    fig = plt.figure(figsize=(16, 10))

    # 1. Distrib montant
    ax1 = plt.subplot(2, 3, 1)
    df['montant_total'].hist(bins=50, edgecolor='black', ax=ax1)
    ax1.set_xlabel('Montant total')
    ax1.set_ylabel('Frequence')
    ax1.set_title('Distribution Montant')
    med1 = df['montant_total'].median()
    ax1.axvline(med1, color='red', linestyle='--', label='Mediane')
    ax1.legend()

    # 2. Distrib nb retraits
    ax2 = plt.subplot(2, 3, 2)
    df['nb_total_de_retraits'].hist(bins=50, edgecolor='black', ax=ax2)
    ax2.set_xlabel('Nb retraits')
    ax2.set_ylabel('Frequence')
    ax2.set_title('Distribution Volume')
    med2 = df['nb_total_de_retraits'].median()
    ax2.axvline(med2, color='red', linestyle='--', label='Mediane')
    ax2.legend()

    # 3. Montant moyen
    ax3 = plt.subplot(2, 3, 3)
    if 'montant_moyen_retrait' in df.columns:
        df['montant_moyen_retrait'].hist(bins=50, edgecolor='black', ax=ax3)
        ax3.set_xlabel('Montant moyen')
        ax3.set_ylabel('Frequence')
        ax3.set_title('Distribution Montant Moyen')

    # 4. Boxplot par type
    ax4 = plt.subplot(2, 3, 4)
    if 'type_carte' in df.columns:
        df.boxplot(column='montant_total', by='type_carte', ax=ax4)
        ax4.set_xlabel('Type')
        ax4.set_ylabel('Montant')
        ax4.set_title('Montant par Type')

    # 5. Scatter
    ax5 = plt.subplot(2, 3, 5)
    ax5.scatter(df['nb_total_de_retraits'], df['montant_total'], alpha=0.5, s=30)
    ax5.set_xlabel('Nb retraits')
    ax5.set_ylabel('Montant')
    ax5.set_title('Montant vs Volume')
    ax5.grid(True, alpha=0.3)

    # 6. Pie types
    ax6 = plt.subplot(2, 3, 6)
    if 'type_carte' in df.columns:
        df['type_carte'].value_counts().plot(kind='pie', ax=ax6, autopct='%1.1f%%')
        ax6.set_ylabel('')
        ax6.set_title('Repartition Types')

    plt.suptitle('Dashboard Analyse DAB', fontsize=16, y=1.00)
    plt.tight_layout()
    plt.show()
```
