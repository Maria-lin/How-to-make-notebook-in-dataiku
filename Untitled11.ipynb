```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "IdentifiantRequeteVQR": range(1, 11),
    "Circuit": ["RVM", "On Us", "RVM", "On Us", "RVM", "RVM", "On Us", "RVM", "On Us", "RVM"],
    "Motif": [None, "Erreur F001", None, "Erreur F004", "Erreur F002", None, "Erreur F003", "Erreur F001", None, "Erreur F004"],
    "Statut": ["Match", "NOAP", "Almost match", "No match", "Match", "NOAP", "No match", "Almost match", "Match", "NOAP"],
    # ICI : génération sur 10 jours
    "TimestampReceptionVop": pd.date_range("2025-09-01", periods=10, freq="D"),
    "DelaiTraitement": np.random.randint(50, 500, size=10),
    "BicDonneurOrdre": ["BNPAFRPP", "SOGEFRPP", "LPSOFPP", "BNPAFRPP", "SOGEFRPP", "LPSOFPP", "BNPAFRPP", "SOGEFRPP", "LPSOFPP", "BNPAFRPP"],
    "BicTeneurCompte": ["CCBPFRPP", "CEPAFRPP", "LPSOFPP", "CEPAFRPP", "CCBPFRPP", "LPSOFPP", "CCBPFRPP", "CEPAFRPP", "LPSOFPP", "CCBPFRPP"],
    "DelaiRVM": np.random.randint(20, 300, size=10),
    "Canal": ["BEL", "APP", "BEL", "APP", "BEL", "APP", "BEL", "APP", "BEL", "APP"]
})

# Extraire uniquement la date (jour)
df["DateReceptionVop"] = df["TimestampReceptionVop"].dt.date

df
```

```python
df["DateReceptionVop"] = pd.to_datetime(df["DateReceptionVop"])

df_rvm = df[df["Circuit"] == "RVM"].copy()

# Définir les 5 derniers jours (par rapport à hier)
end_date = date.today() - timedelta(days=1)
start_date = end_date - timedelta(days=4)
last5days = pd.date_range(start=start_date, end=end_date)

# Fonction pour tracer un graphe par canal
def plot_requetes_par_statut(df, canal):
    subset = df[(df["Canal"] == canal) &
                (df["DateReceptionVop"].dt.date >= start_date) &
                (df["DateReceptionVop"].dt.date <= end_date)]

    # Compter les requêtes par date et statut
    counts = subset.groupby([subset["DateReceptionVop"].dt.date, "Statut"]).size().reset_index(name="Nombre")

    # S'assurer que toutes les dates/statuts apparaissent (même si 0)
    all_combinations = pd.MultiIndex.from_product(
        [last5days.date, df["Statut"].unique()],
        names=["DateReceptionVop", "Statut"]
    )
    counts = counts.set_index(["DateReceptionVop", "Statut"]).reindex(all_combinations, fill_value=0).reset_index()

    # ---------------------------
    # Plot
    # ---------------------------
    plt.figure(figsize=(10,6))
    sns.barplot(data=counts, x="DateReceptionVop", y="Nombre", hue="Statut")

    plt.title(f"Nombre de requêtes VOP par statut\nCircuit RVM – Canal {canal} (5 derniers jours)")
    plt.xlabel("Date de réception")
    plt.ylabel("Nombre de requêtes")
    plt.legend(title="Statut")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# ---------------------------
# Générer les 2 graphes
# ---------------------------
plot_requetes_par_statut(df_rvm, "BEL")
plot_requetes_par_statut(df_rvm, "APP")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, timedelta

# ---------------------------
# ⚠️ Exemple simulé (remplace avec ton dataset réel)
# ---------------------------
data = {
    "date_reception_vop": pd.date_range(end=date.today(), periods=10).tolist() * 4,
    "Circuit": ["RVM"] * 20 + ["On Us"] * 20,
    "canal": (["BEL"] * 10 + ["APP"] * 10) * 2,
    "statut": (["MTCH", "CMTCCH", "NMTC", "NOAP"] * 10)
}
df = pd.DataFrame(data)

# ---------------------------
# S’assurer que la colonne date est bien en datetime
# ---------------------------
df["date_reception_vop"] = pd.to_datetime(df["date_reception_vop"])

# ---------------------------
# 👉 Tu fixes ici tes dates manuellement
# ---------------------------
end_date = date(2025, 9, 23)   # à modifier quand tu veux
start_date = end_date - timedelta(days=4)  # 5 derniers jours

# ---------------------------
# Fonction générique pour tracer un graphe avec légende à l’extérieur
# ---------------------------
def plot_requetes_par_statut(df, circuit, canal):
    subset = df[
        (df["Circuit"] == circuit) &
        (df["canal"] == canal) &
        (df["date_reception_vop"].dt.date >= start_date) &
        (df["date_reception_vop"].dt.date <= end_date)
    ]

    # Compter les requêtes par date et statut
    counts = subset.groupby([subset["date_reception_vop"].dt.date, "statut"]).size().reset_index(name="Nombre")

    # S’assurer que toutes les dates/statuts apparaissent (même si 0)
    all_combinations = pd.MultiIndex.from_product(
        [pd.date_range(start=start_date, end=end_date).date, df["statut"].unique()],
        names=["date_reception_vop", "statut"]
    )
    counts = counts.set_index(["date_reception_vop", "statut"]).reindex(all_combinations, fill_value=0).reset_index()

    # ---------------------------
    # Plot
    # ---------------------------
    plt.figure(figsize=(10,6))
    sns.barplot(data=counts, x="date_reception_vop", y="Nombre", hue="statut")

    plt.title(f"Nombre de requêtes VOP par statut\nCircuit {circuit} – Canal {canal} (5 jours)")
    plt.xlabel("Date de réception")
    plt.ylabel("Nombre de requêtes")

    # 🔹 Légende à l’extérieur du graphe
    plt.legend(
        title="Statut",
        bbox_to_anchor=(1.05, 1),  # position sur X et Y
        loc='upper left',          # ancrage
        borderaxespad=0.
    )

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# ---------------------------
# Graphe Requête 1 : Circuit RVM
# ---------------------------
plot_requetes_par_statut(df, "RVM", "BEL")
plot_requetes_par_statut(df, "RVM", "APP")

# ---------------------------
# Graphe Requête 2 : Circuit On Us
# ---------------------------
plot_requetes_par_statut(df, "On Us", "BEL")
plot_requetes_par_statut(df, "On Us", "APP")

# ---------------------------
# Pancarte explicative (hors graphe, en console)
# ---------------------------
print("📝 Correspondance des statuts :")
print(" - MTCH   = Match")
print(" - CMTCCH = Almost match")
print(" - NMTC   = No match")
print(" - NOAP   = VOP impossible")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, timedelta

# ---------------------------
# Exemple de données avec motifs
# ---------------------------
data = {
    "date_reception_vop": [
        "2025.06.19.T0000000","2025.06.19.T0000000","2025.06.20.T0000000",
        "2025.06.21.T0000000","2025.06.21.T0000000","2025.06.21.T0000000",
        "2025.06.19.T0000000","2025.06.19.T0000000","2025.06.20.T0000000",
        "2025.06.21.T0000000","2025.06.21.T0000000","2025.06.21.T0000000"
    ],
    "Circuit": ["RVM"]*12,
    "canal": ["BELL","BELL","BELL","BELL","BELL","BELL","HAP","HAP","HAP","HAP","HAP","HAP"],
    "statut": ["NOAP","NOAP","MTCH","NOAP","MTCH","NMTC","NOAP","NOAP","MTCH","NOAP","MTCH","NMTC"],
    "motif": ["Motif1","Motif2","Motif1","Motif1","Motif2","Motif3","Motif1","Motif2","Motif1","Motif2","Motif3","Motif1"]
}

df = pd.DataFrame(data)

# ---------------------------
# Nettoyage de la colonne date
# ---------------------------
df["date_reception_vop_clean"] = (
    df["date_reception_vop"]
      .str.replace(r"\.T.*", "", regex=True)
      .str.replace(r"\.", "-", regex=True)
)
df["date_reception_vop_clean"] = pd.to_datetime(df["date_reception_vop_clean"], format="%Y-%m-%d")

# ---------------------------
# Période analysée (2 derniers jours)
# ---------------------------
end_date = df["date_reception_vop_clean"].max().date()
start_date = end_date - timedelta(days=1)

# ---------------------------
# Fonction : nombre de requêtes NOAP par motif
# ---------------------------
def plot_noap_par_motif(df, canal):
    subset = df[
        (df["Circuit"] == "RVM") &
        (df["canal"] == canal) &
        (df["statut"] == "NOAP") &
        (df["date_reception_vop_clean"].dt.date >= start_date) &
        (df["date_reception_vop_clean"].dt.date <= end_date)
    ]

    # Compter par date et motif
    counts = subset.groupby([subset["date_reception_vop_clean"].dt.date, "motif"]).size().reset_index(name="Nombre")

    # S’assurer que toutes les dates et motifs existent
    all_combinations = pd.MultiIndex.from_product(
        [pd.date_range(start=start_date, end=end_date).date, df["motif"].unique()],
        names=["date_reception_vop_clean", "motif"]
    )
    counts = counts.set_index(["date_reception_vop_clean", "motif"]).reindex(all_combinations, fill_value=0).reset_index()

    # Graphe
    plt.figure(figsize=(8,5))
    sns.barplot(data=counts, x="date_reception_vop_clean", y="Nombre", hue="motif")
    plt.title(f"Nombre de requêtes NOAP par motif - Circuit RVM - Canal {canal}")
    plt.xlabel("Date de réception")
    plt.ylabel("Nombre de requêtes")
    plt.xticks(rotation=45)
    plt.legend(title="Motifs", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# ---------------------------
# Affichage des graphiques pour BELL et HAP
# ---------------------------
for canal in ["BELL", "APP"]:
    plot_noap_par_motif(df, canal)
```
