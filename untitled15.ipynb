```python
# -------------------------------
# Import nécessaire (si pas déjà importé)
# -------------------------------
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import plotly.express as px

# -------------------------------
# Création du jeu de données de test
# -------------------------------
dates = pd.date_range(datetime.today() - timedelta(days=7), periods=8)
bic_donneur_ordre_options = ['PSSTFPPXXX']  # LBP
bic_teneur_compte_options = ['BIC001', 'BIC002', 'BIC003', 'BIC004', 'BIC005',
                             'BIC006', 'BIC007', 'BIC008', 'BIC009', 'BIC010', 'BIC011']
statut_options = ['STATUT_A', 'STATUT_B', 'STATUT_C', 'STATUT_D']

data = []
np.random.seed(42)

for date in dates:
    for bic_teneur in bic_teneur_options:
        nb_reponses = np.random.randint(5, 50)
        for _ in range(nb_reponses):
            data.append({
                'date': date,
                'bic_donneur_ordre': 'PSSTFPPXXX',
                'bic_teneur_compte': bic_teneur,
                'statut': np.random.choice(statut_options)
            })

df_test = pd.DataFrame(data)
print(f"Total lignes dans le dataset de test : {df_test.shape[0]}")

# -------------------------------
# Définition des fonctions
# -------------------------------
def get_date_range(days_back=8):
    end_date = datetime.today()
    start_date = end_date - timedelta(days=days_back)
    return start_date, end_date

def filter_by_date(df, start_date, end_date):
    df['date'] = pd.to_datetime(df['date'])
    return df[(df['date'] >= start_date) & (df['date'] <= end_date)]

def tableau_par_bic_statut(df, start_date, end_date, bic_lbp='PSSTFPPXXX'):
    df_filtered = filter_by_date(df, start_date, end_date)
    df_filtered = df_filtered[df_filtered['bic_donneur_ordre'] == bic_lbp]
    grouped = df_filtered.groupby(['date', 'bic_teneur_compte', 'statut']).size().reset_index(name='nb_reponse')
    total_bic = df_filtered.groupby(['date', 'bic_teneur_compte']).size().reset_index(name='nbtotal_reponse')
    merged = grouped.merge(total_bic, on=['date', 'bic_teneur_compte'])
    merged['pourcentage'] = (merged['nb_reponse'] / merged['nbtotal_reponse'] * 100).round(2)
    total_confreres = df_filtered.groupby('date').size().reset_index(name='nbtotal_reponse_tous')
    final = merged.merge(total_confreres, on='date')
    return final.sort_values(['date', 'bic_teneur_compte', 'statut'])

def top10_confrere_par_date(df, start_date, end_date, bic_lbp='PSSTFPPXXX'):
    df_filtered = filter_by_date(df, start_date, end_date)
    df_filtered = df_filtered[df_filtered['bic_teneur_compte'] != bic_lbp]
    total_confreres = df_filtered.groupby(['date', 'bic_teneur_compte']).size().reset_index(name='nb_total_reponse_conf')
    top10 = total_confreres.groupby('date').apply(lambda x: x.nlargest(10, 'nb_total_reponse_conf')).reset_index(drop=True)
    total_tous = df_filtered.groupby('date').size().reset_index(name='nb_total_reponse_tous')
    return top10.merge(total_tous, on='date').sort_values(['date', 'nb_total_reponse_conf'], ascending=[True, False])

def top10_confrere_global(df, start_date, end_date, bic_lbp='PSSTFPPXXX'):
    df_filtered = filter_by_date(df, start_date, end_date)
    df_filtered = df_filtered[df_filtered['bic_teneur_compte'] != bic_lbp]
    total_confreres = df_filtered.groupby('bic_teneur_compte').size().reset_index(name='nb_total_reponse_conf')
    top10_global = total_confreres.nlargest(10, 'nb_total_reponse_conf')
    top10_global['nb_total_reponse_tous'] = df_filtered.shape[0]
    return top10_global.sort_values('nb_total_reponse_conf', ascending=False)

def plot_pourcentage_confrere(df, start_date, end_date, bic_lbp='PSSTFPPXXX'):
    df_tableau = tableau_par_bic_statut(df, start_date, end_date, bic_lbp)
    fig = px.line(df_tableau,
                  x='date',
                  y='pourcentage',
                  color='bic_teneur_compte',
                  title='Pourcentage de réponses par confrère par date',
                  markers=True,
                  template='plotly_white')
    fig.update_layout(xaxis_title='Date', yaxis_title='Pourcentage (%)', legend_title='Confrère')
    fig.show()

def plot_statut(df, start_date, end_date, bic_lbp='PSSTFPPXXX'):
    df_tableau = tableau_par_bic_statut(df, start_date, end_date, bic_lbp)
    fig_line = px.line(df_tableau,
                       x='date', y='pourcentage', color='statut', markers=True,
                       title='Pourcentage de réponses par statut par date', template='plotly_white')
    fig_line.update_layout(xaxis_title='Date', yaxis_title='Pourcentage (%)', legend_title='Statut')
    fig_line.show()
    fig_bar = px.bar(df_tableau,
                     x='date', y='pourcentage', color='statut', barmode='stack',
                     title='Pourcentage de réponses par statut par date (empilé)', template='plotly_white')
    fig_bar.update_layout(xaxis_title='Date', yaxis_title='Pourcentage (%)', legend_title='Statut')
    fig_bar.show()

# -------------------------------
# Test des fonctions
# -------------------------------
start_date, end_date = get_date_range(days_back=8)

# 1 - Tableau détaillé
tableau = tableau_par_bic_statut(df_test, start_date, end_date)
print("\n--- Tableau détaillé par BIC et statut ---")
print(tableau.head())

# 2 - Top 10 par date
top10_date = top10_confrere_par_date(df_test, start_date, end_date)
print("\n--- Top 10 confrères par date ---")
print(top10_date.head(20))

# 3 - Top 10 global
top10_global = top10_confrere_global(df_test, start_date, end_date)
print("\n--- Top 10 confrères global ---")
print(top10_global)

# 4 - Plot linéaire pourcentage par confrère
plot_pourcentage_confrere(df_test, start_date, end_date)

# 5 - Plot moderne par statut
plot_statut(df_test, start_date, end_date)
```

```python
def tableau_temps_traitement_counts(df, start_date, end_date):
    """
    Prépare un tableau complet par jour et par circuit
    avec le nombre de requêtes et le temps moyen de traitement.

    Arguments :
        df : DataFrame avec ['TimestampReceptionVOP','Circuit','DelaiTraitement','IdentifiantRequeteVOP']
        start_date, end_date : YYYY-MM-DD ou datetime
    """
    # Normalisation des dates
    df['date_col'] = pd.to_datetime(df['TimestampReceptionVOP'], errors='coerce').dt.normalize()
    start_date = pd.to_datetime(start_date).normalize()
    end_date = pd.to_datetime(end_date).normalize()

    # Filtrage période
    df_filtre = df[(df['date_col'] >= start_date) & (df['date_col'] <= end_date)].copy()

    # Agrégation par date et circuit
    df_counts = (
        df_filtre.groupby(['date_col','Circuit'])
        .agg(
            nb_requetes=('IdentifiantRequeteVOP','count'),
            delai_moyen=('DelaiTraitement','mean')
        )
        .reset_index()
    )

    # Compléter toutes les dates × circuits
    all_dates = pd.date_range(start=start_date, end=end_date, freq='D')
    all_circuits = df['Circuit'].unique()
    idx = pd.MultiIndex.from_product([all_dates, all_circuits],
                                     names=['date_col','Circuit'])
    df_counts = df_counts.set_index(['date_col','Circuit']).reindex(idx, fill_value=0).reset_index()

    # Trier par date
    df_counts = df_counts.sort_values('date_col').reset_index(drop=True)
    return df_counts
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_delai_traitement_par_circuit(df, circuit, start_date, end_date):
    """
    Trace le temps moyen de traitement par jour pour un circuit donné.

    Arguments :
        df : DataFrame contenant ['TimestampReceptionVOP','Circuit','DelaiTraitement','IdentifiantRequeteVOP']
        circuit : 'RVM' ou 'On Us'
        start_date, end_date : YYYY-MM-DD ou datetime
    """
    # Normalisation de la date
    df['date_col'] = pd.to_datetime(df['TimestampReceptionVOP'], errors='coerce').dt.normalize()
    start_date = pd.to_datetime(start_date).normalize()
    end_date = pd.to_datetime(end_date).normalize()

    # Filtrage par circuit et plage de dates
    subset = df[
        (df['Circuit'] == circuit) &
        (df['date_col'] >= start_date) &
        (df['date_col'] <= end_date)
    ]

    # Agrégation : calcul du temps moyen par jour
    counts = (
        subset.groupby('date_col')['DelaiTraitement']
        .mean()
        .reset_index(name='delai_moyen')
    )

    # Compléter toutes les dates même si pas de données
    all_dates = pd.date_range(start=start_date, end=end_date)
    counts = counts.set_index('date_col').reindex(all_dates, fill_value=0).reset_index()
    counts.rename(columns={'index':'date_col'}, inplace=True)

    # Plot
    plt.figure(figsize=(10,5))
    sns.lineplot(data=counts, x='date_col', y='delai_moyen', marker='o')
    plt.title(f"Temps moyen de traitement par jour - Circuit = {circuit}")
    plt.xlabel("Date")
    plt.ylabel("Temps moyen de traitement")

    ax = plt.gca()
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gcf().autofmt_xdate()

    plt.tight_layout()
    plt.show()
plot_delai_traitement_par_circuit(df, "RVM", "2025-09-24", "2025-09-27")
plot_delai_traitement_par_circuit(df, "On Us", "2025-09-24", "2025-09-27")
```

```python
def top10_confreres(df):
    """
    Retourne un tableau avec les 10 confrères ayant le plus grand volume total.
    Une ligne par confrère, avec total et pourcentage.
    """
    # Total par confrère
    total_par_confrere = df.groupby("bic_teneur_compte")["nb"].sum()

    # Top 10
    top10 = total_par_confrere.nlargest(10)

    # Créer un tableau final avec total et pourcentage global
    df_top10 = pd.DataFrame({
        "bic_teneur_compte": top10.index,
        "total": top10.values
    })
    df_top10["pourcentage_global"] = (df_top10["total"] / df_top10["total"].sum()) * 100

    return df_top10
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date, timedelta

# ----------------------------
# 1. Création d'un mini jeu de données
# ----------------------------
np.random.seed(42)  # reproductible
jours = [date.today() - timedelta(days=i) for i in range(10)]

data = []
for d in jours:
    for _ in range(np.random.randint(8, 15)):  # lignes par jour
        bic_donneur = "PSSTFRPPXXX" if np.random.rand() > 0.2 else "AUTREBIC123"
        bic_teneur = "AUTREBIC123" if bic_donneur == "PSSTFRPPXXX" else "PSSTFRPPXXX"
        statut = np.random.choice(["NOAP", "ACCP", "RJCT"], p=[0.3, 0.5, 0.2])
        motif = "FOO4" if (statut == "NOAP" and np.random.rand() > 0.5) else "AUTRE"
        data.append([d, bic_donneur, bic_teneur, statut, motif])

df = pd.DataFrame(data, columns=["date", "bic_donneur_ordre", "bic_teneur_compte", "statut", "motif"])


# ----------------------------
# 2. Fonction de calcul tableau
# ----------------------------
def calcul_vop_impossible(df):
    today = date.today()
    last_days = [today - timedelta(days=i) for i in range(7, -1, -1)]  # 8 derniers jours (ordonnés)

    # Filtre sur les conditions
    df_filtre = df[(df["bic_donneur_ordre"] == "PSSTFRPPXXX") &
                   (df["bic_teneur_compte"] != "PSSTFRPPXXX") &
                   (df["date"].isin(last_days))]

    results = []
    for d in last_days:
        df_day = df_filtre[df_filtre["date"] == d]
        total = len(df_day)
        vop = len(df_day[(df_day["statut"] == "NOAP") & (df_day["motif"] == "FOO4")])
        pct = (vop / total * 100) if total > 0 else 0
        results.append([d, round(pct, 2), total])

    stats = pd.DataFrame(results, columns=["date", "pourcentage", "nb_total"])
    return stats


# ----------------------------
# 3. Fonction de visualisation
# ----------------------------
def plot_vop(stats):
    plt.figure(figsize=(8,4))
    plt.plot(stats["date"], stats["pourcentage"], marker="o", linestyle="-", color="blue")
    plt.title("Pourcentage de réponses VOP impossibles (8 derniers jours)")
    plt.xlabel("Date")
    plt.ylabel("Pourcentage (%)")
    plt.xticks(rotation=45)
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()


# ----------------------------
# 4. Exemple d'utilisation
# ----------------------------
stats = calcul_vop_impossible(df)
print(stats)   # affiche le tableau
plot_vop(stats)
```

```python
import pandas as pd
import plotly.express as px
from datetime import date, timedelta

# --- Dates de filtrage ---
end_date = date.today()
start_date = end_date - timedelta(days=8)

# --- 0. Filtrer le dataframe par date ---
def filter_by_date(df, date_col='date'):
    """Filtre le dataframe entre start_date et end_date inclus."""
    df[date_col] = pd.to_datetime(df[date_col])
    mask = (df[date_col] >= pd.to_datetime(start_date)) & (df[date_col] <= pd.to_datetime(end_date))
    return df.loc[mask].copy()

# --- 1. Tableau par bic_teneur_compte et statut ---
def table_by_status(df, bic_lbp='PSSTFPPXXX'):
    """
    Retourne un tableau :
    date | bic_teneur_compte | statut | nb_reponse | nb_total_reponse_confrere | pourcentage | nb_total_reponses_tous_confreres
    """
    df_filtered = df.copy()

    # Nombre total de réponses par confrère
    total_per_confrere = df_filtered.groupby(['date', 'bic_teneur_compte']).size().reset_index(name='nb_total_reponse_confrere')

    # Nombre de réponses par statut
    status_count = df_filtered.groupby(['date', 'bic_teneur_compte', 'statut']).size().reset_index(name='nb_reponse')

    # Merge pour total par confrère
    merged = status_count.merge(total_per_confrere, on=['date', 'bic_teneur_compte'])

    # Pourcentage par statut
    merged['pourcentage'] = merged['nb_reponse'] / merged['nb_total_reponse_confrere'] * 100

    # Total réponses tous confrères
    total_all = df_filtered.groupby('date').size().reset_index(name='nb_total_reponses_tous_confreres')
    merged = merged.merge(total_all, on='date')

    # Trier par bic_teneur_compte
    merged = merged.sort_values(['date', 'bic_teneur_compte'])

    return merged

# --- 2. Top 10 confrères par date ---
def top10_confreres_per_date(df):
    """
    Retourne pour chaque date le top 10 confrères par volumétrie.
    date | bic_teneur_compte | nb_total_reponse_confrere | nb_total_reponses_tous_confreres
    """
    df_total = df.groupby(['date', 'bic_teneur_compte']).size().reset_index(name='nb_total_reponse_confrere')

    # Total réponses par date
    total_all = df.groupby('date').size().reset_index(name='nb_total_reponses_tous_confreres')
    df_total = df_total.merge(total_all, on='date')

    # Top 10 par date
    df_total['rank'] = df_total.groupby('date')['nb_total_reponse_confrere'].rank(method='first', ascending=False)
    top10 = df_total[df_total['rank'] <= 10].drop(columns='rank').sort_values(['date', 'nb_total_reponse_confrere'], ascending=[True, False])

    return top10

# --- 3. Top 10 confrères global ---
def top10_confreres_global(df):
    """
    Retourne le top 10 confrères global par volumétrie sur toute la période.
    bic_teneur_compte | nb_total_reponse_confrere | nb_total_reponses_tous_confreres
    """
    df_total = df.groupby('bic_teneur_compte').size().reset_index(name='nb_total_reponse_confrere')
    total_all = df.shape[0]
    df_total['nb_total_reponses_tous_confreres'] = total_all

    top10 = df_total.sort_values('nb_total_reponse_confrere', ascending=False).head(10)
    return top10

# --- 4. Plot linéaire : pourcentage de réponses par confrère par date ---
import plotly.express as px

def plot_percentage_by_confrere(df):
    """
    Graphique linéaire interactif du pourcentage de réponses par confrère par date.
    """
    df_plot = table_by_status(df)  # On réutilise la fonction précédente

    # Résumé par date et confrère
    df_summary = df_plot.groupby(['date', 'bic_teneur_compte']).agg({
        'nb_reponse':'sum',
        'nb_total_reponse_confrere':'sum'
    }).reset_index()
    df_summary['pourcentage'] = df_summary['nb_reponse'] / df_summary['nb_total_reponse_confrere'] * 100

    # Création du graphique
    fig = px.line(
        df_summary,
        x='date',
        y='pourcentage',
        color='bic_teneur_compte',
        title='Pourcentage de réponses par confrère par date'
    )

    # Ajouter les marqueurs (points)
    fig.update_traces(mode='lines+markers')

    fig.show()


def plot_percentage_by_status(df):
    """
    Génère 2 plots interactifs :
    1. Ligne : pourcentage total par date
    2. Barres empilées : pourcentage par statut par date
    """
    df_plot = table_by_status(df)

    # Pivot pour bar empilé
    pivot_status = df_plot.pivot_table(
        index='date',
        columns='statut',
        values='pourcentage',
        aggfunc='sum'
    ).reset_index()

    # Graphique ligne : pourcentage total par date
    df_total_percentage = df_plot.groupby('date')['pourcentage'].sum().reset_index()
    fig_line = px.line(
        df_total_percentage,
        x='date',
        y='pourcentage',
        title='Pourcentage total de réponses par date'
    )
    fig_line.update_traces(mode='lines+markers')  # Ajoute les marqueurs sur la ligne

    # Graphique bar empilé
    fig_bar = px.bar(
        pivot_status,
        x='date',
        y=pivot_status.columns[1:],
        title='Pourcentage de réponses par statut',
        labels={'value':'Pourcentage'},
        barmode='stack'  # Pour empiler les statuts
    )

    fig_line.show()
    fig_bar.show()

import pandas as pd
from datetime import datetime, timedelta

# Création du DataFrame factice
data = {
    'date': [
        datetime.today() - timedelta(days=7),
        datetime.today() - timedelta(days=7),
        datetime.today() - timedelta(days=6),
        datetime.today() - timedelta(days=6),
        datetime.today() - timedelta(days=5),
        datetime.today() - timedelta(days=5),
        datetime.today() - timedelta(days=4),
        datetime.today() - timedelta(days=4),
        datetime.today() - timedelta(days=3),
        datetime.today() - timedelta(days=3),
    ],
    'bic_teneur_compte': [
        'PSSTFPPXXX', 'BNPAFRPPXXX', 'PSSTFPPXXX', 'BNPAFRPPXXX',
        'CICDFRPPXXX', 'PSSTFPPXXX', 'BNPAFRPPXXX', 'CICDFRPPXXX',
        'PSSTFPPXXX', 'BNPAFRPPXXX'
    ],
    'statut': [
        'OK', 'KO', 'OK', 'KO', 'OK', 'KO', 'OK', 'KO', 'OK', 'KO'
    ]
}

df = pd.DataFrame(data)

# Affichage du DataFrame
print("=== Jeu de données factice ===")
print(df)


df_filtered = filter_by_date(df)
table_status = table_by_status(df_filtered)
top10_date = top10_confreres_per_date(df_filtered)
top10_global = top10_confreres_global(df_filtered)
plot_percentage_by_confrere(df_filtered)
plot_percentage_by_status(df_filtered)
```

```python
import pandas as pd
import plotly.express as px

def plot_percentage_by_status(df):
    """
    Génère 2 plots interactifs :
    1. Ligne : pourcentage par statut par date (nb réponses statut / nb total réponses date)
    2. Barres empilées : pourcentage par statut par date
    """
    # S'assurer que la colonne date est datetime
    df['date'] = pd.to_datetime(df['date'])

    # Total de réponses par date
    total_per_date = df.groupby('date').size().reset_index(name='nb_total')

    # Total par date et statut
    df_status = df.groupby(['date', 'statut']).size().reset_index(name='nb_statut')

    # Merge pour calculer le pourcentage
    df_status = df_status.merge(total_per_date, on='date')
    df_status['pourcentage'] = df_status['nb_statut'] / df_status['nb_total'] * 100

    # --- Plot ligne : chaque statut devient une ligne ---
    fig_line = px.line(
        df_status,
        x='date',
        y='pourcentage',
        color='statut',
        markers=True,
        title='Pourcentage de réponses par statut par date'
    )

    # --- Pivot pour barres empilées ---
    pivot_status = df_status.pivot_table(
        index='date',
        columns='statut',
        values='pourcentage',
        aggfunc='sum'
    ).reset_index()

    fig_bar = px.bar(
        pivot_status,
        x='date',
        y=pivot_status.columns[1:],  # toutes les colonnes statut
        title='Pourcentage de réponses par statut (barres empilées)',
        labels={'value':'Pourcentage'},
        barmode='stack'
    )

    fig_line.show()
    fig_bar.show()
```

```python
import pandas as pd
from datetime import datetime, timedelta

# --- Fonction 1 : Tableau par confrère / statut / total du confrère ---
def table_par_confrere_statut(df, end_date):
    df_filtered = df[
        (df['bic_donneur_ordre'] == 'PPSSTFRPPXXX') &
        (df['bic_teneur_compte'] != 'PPSSTFRPPXXX') &
        (df['date'] == end_date)
    ].copy()

    # Nombre de réponses par confrère / statut
    status_count = df_filtered.groupby(['bic_teneur_compte','statut']).size().reset_index(name='nb_reponse')

    # Total de réponses par confrère
    total_per_confrere = df_filtered.groupby('bic_teneur_compte').size().reset_index(name='nb_total_reponse_confrere')

    # Merge pour ajouter total par confrère
    tableau1 = status_count.merge(total_per_confrere, on='bic_teneur_compte')

    # Pourcentage par statut sur total du confrère
    tableau1['pct_confrere'] = (tableau1['nb_reponse'] / tableau1['nb_total_reponse_confrere'] * 100).round(2)

    # Colonnes claires
    tableau1 = tableau1.rename(columns={
        'bic_teneur_compte': 'Confrère',
        'statut': 'Statut',
        'nb_reponse': 'Nb réponses',
        'nb_total_reponse_confrere': 'Total réponses confrère',
        'pct_confrere': '% réponses / confrère'
    })

    return tableau1

# --- Fonction 2 : Tableau par confrère / statut / total de tous les confrères ---
def table_general(df, end_date):
    df_filtered = df[
        (df['bic_donneur_ordre'] == 'PPSSTFRPPXXX') &
        (df['bic_teneur_compte'] != 'PPSSTFRPPXXX') &
        (df['date'] == end_date)
    ].copy()

    # Nombre de réponses par confrère / statut
    status_count = df_filtered.groupby(['bic_teneur_compte','statut']).size().reset_index(name='nb_reponse')

    # Total de réponses tous confrères
    total_all = df_filtered.shape[0]

    # Ajout du total et calcul du pourcentage global
    status_count['Total réponses tous confrères'] = total_all
    status_count['% réponses / tous confrères'] = (status_count['nb_reponse'] / total_all * 100).round(2)

    # Colonnes claires
    status_count = status_count.rename(columns={
        'bic_teneur_compte': 'Confrère',
        'statut': 'Statut',
        'nb_reponse': 'Nb réponses'
    })

    return status_count

# --- Fonction pour afficher les tableaux de façon moderne ---
def display_tableau_modern(tableau, pourcentage_col, titre):
    styled = tableau.style.background_gradient(
        subset=[pourcentage_col], cmap='Blues'
    ).format({
        pourcentage_col: "{:.2f} %"
    }).set_caption(titre)\
      .set_table_styles([{'selector': 'th', 'props': [('background-color', '#40466e'),
                                                      ('color', 'white'),
                                                      ('font-size', '12pt')]}])
    return styled

# --- Exemple d'utilisation ---
end_date = datetime.today() - timedelta(days=1)

# Création des tableaux
tableau1 = table_par_confrere_statut(df, end_date)
tableau2 = table_general(df, end_date)

# Titres formels pour les tableaux
titre1 = "Pourcentage de réponses des confrères par statut sur le total des réponses de ce confrère"
titre2 = "Pourcentage de réponses des confrères par statut sur le total de tous les confrères"

# Affichage moderne
display(display_tableau_modern(tableau1, '% réponses / confrère', titre1))
display(display_tableau_modern(tableau2, '% réponses / tous confrères', titre2))
```

```python
import plotly.express as px

def plot_pourcentage_tous_confreres(df, top10_df):
    """
    Plot en barres empilées : pourcentage par statut par confrère
    - % calculé sur le total de réponses de tous les confrères
    """
    # Garder seulement les BIC du top 10
    df_filtered = df[df['bic_teneur_compte'].isin(top10_df['bic_teneur_compte'])].copy()

    # Compter le nombre de réponses par confrère et statut
    df_grouped = df_filtered.groupby(['bic_teneur_compte','statut']).size().reset_index(name='nb_reponse')

    # Total de réponses de tous les confrères
    total_all = df_grouped['nb_reponse'].sum()

    # Calcul du pourcentage
    df_grouped['pourcentage'] = df_grouped['nb_reponse'] / total_all * 100

    # Plot empilé
    fig = px.bar(
        df_grouped,
        x='bic_teneur_compte',
        y='pourcentage',
        color='statut',
        text=df_grouped['pourcentage'].round(2).astype(str) + '%',
        title='Pourcentage de réponses par statut (sur total de tous les confrères)'
    )

    fig.update_traces(textposition='inside')
    fig.update_layout(barmode='stack', xaxis_title='Confrère', yaxis_title='Pourcentage (%)')

    fig.show()


def plot_pourcentage_par_confrere(df, top10_df):
    """
    Plot en barres empilées : pourcentage par statut par confrère
    - % calculé sur le total de réponses du confrère
    """
    # Garder seulement les BIC du top 10
    df_filtered = df[df['bic_teneur_compte'].isin(top10_df['bic_teneur_compte'])].copy()

    # Compter le nombre de réponses par confrère et statut
    df_grouped = df_filtered.groupby(['bic_teneur_compte','statut']).size().reset_index(name='nb_reponse')

    # Total par confrère
    total_per_confrere = df_grouped.groupby('bic_teneur_compte')['nb_reponse'].transform('sum')

    # Calcul du pourcentage
    df_grouped['pourcentage'] = df_grouped['nb_reponse'] / total_per_confrere * 100

    # Plot empilé
    fig = px.bar(
        df_grouped,
        x='bic_teneur_compte',
        y='pourcentage',
        color='statut',
        text=df_grouped['pourcentage'].round(2).astype(str) + '%',
        title='Pourcentage de réponses par statut (sur total du confrère)'
    )

    fig.update_traces(textposition='inside')
    fig.update_layout(barmode='stack', xaxis_title='Confrère', yaxis_title='Pourcentage (%)')

    fig.show()
```
