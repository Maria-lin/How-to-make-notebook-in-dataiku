```python
df["DateReceptionVop"] = pd.to_datetime(df["DateReceptionVop"])

df_rvm = df[df["Circuit"] == "RVM"].copy()

# Définir les 5 derniers jours (par rapport à hier)
end_date = date.today() - timedelta(days=1)
start_date = end_date - timedelta(days=4)
last5days = pd.date_range(start=start_date, end=end_date)

# Fonction pour tracer un graphe par canal
def plot_requetes_par_statut(df, canal):
    subset = df[(df["Canal"] == canal) &
                (df["DateReceptionVop"].dt.date >= start_date) &
                (df["DateReceptionVop"].dt.date <= end_date)]

    # Compter les requêtes par date et statut
    counts = subset.groupby([subset["DateReceptionVop"].dt.date, "Statut"]).size().reset_index(name="Nombre")

    # S'assurer que toutes les dates/statuts apparaissent (même si 0)
    all_combinations = pd.MultiIndex.from_product(
        [last5days.date, df["Statut"].unique()],
        names=["DateReceptionVop", "Statut"]
    )
    counts = counts.set_index(["DateReceptionVop", "Statut"]).reindex(all_combinations, fill_value=0).reset_index()

    # ---------------------------
    # Plot
    # ---------------------------
    plt.figure(figsize=(10,6))
    sns.barplot(data=counts, x="DateReceptionVop", y="Nombre", hue="Statut")

    plt.title(f"Nombre de requêtes VOP par statut\nCircuit RVM – Canal {canal} (5 derniers jours)")
    plt.xlabel("Date de réception")
    plt.ylabel("Nombre de requêtes")
    plt.legend(title="Statut")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# ---------------------------
# Générer les 2 graphes
# ---------------------------
plot_requetes_par_statut(df_rvm, "BEL")
plot_requetes_par_statut(df_rvm, "APP")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, timedelta

# ---------------------------
# Exemple simulé (remplace avec ton dataset réel)
# ---------------------------
data = {
    "date_reception_vop": pd.date_range(end=date.today(), periods=10).tolist() * 4,
    "Circuit": ["RVM"] * 40,
    "canal": (["BEL"] * 10 + ["APP"] * 10) * 2,
    "statut": ["NOAP"] * 40,  # on met que du NOAP pour l'exemple
    "motifs": (["Motif1","Motif2","Motif3","Motif4","Motif1"] * 8)
}
df = pd.DataFrame(data)

# ---------------------------
# S’assurer que la colonne date est bien en datetime
# ---------------------------
df["date_reception_vop"] = pd.to_datetime(df["date_reception_vop"], errors='coerce')

# ---------------------------
# 👉 Choisir ici la date de fin
# ---------------------------
end_date = date(2025, 9, 23)   # à modifier quand tu veux
start_date = end_date - timedelta(days=4)  # 5 derniers jours

# ---------------------------
# Fonction générique pour tracer un graphe NOAP par motif
# ---------------------------
def plot_noap_par_motif(df, circuit, canal):
    subset = df[
        (df["Circuit"] == circuit) &
        (df["canal"] == canal) &
        (df["statut"] == "NOAP") &
        (df["date_reception_vop"].dt.date >= start_date) &
        (df["date_reception_vop"].dt.date <= end_date)
    ]

    # Compter les requêtes par date et motif
    counts = subset.groupby([subset["date_reception_vop"].dt.date, "motifs"]).size().reset_index(name="Nombre")

    # S’assurer que toutes les dates/motifs apparaissent (même si 0)
    all_combinations = pd.MultiIndex.from_product(
        [pd.date_range(start=start_date, end=end_date).date, df["motifs"].unique()],
        names=["date_reception_vop", "motifs"]
    )
    counts = counts.set_index(["date_reception_vop", "motifs"]).reindex(all_combinations, fill_value=0).reset_index()

    # ---------------------------
    # Plot
    # ---------------------------
    plt.figure(figsize=(10,6))
    sns.barplot(data=counts, x="date_reception_vop", y="Nombre", hue="motifs")

    plt.title(f"Nombre de requêtes NOAP par motif\nCircuit {circuit} – Canal {canal} (5 jours)")
    plt.xlabel("Date de réception")
    plt.ylabel("Nombre de requêtes")

    # 🔹 Légende à l’extérieur du graphe
    plt.legend(
        title="Motif",
        bbox_to_anchor=(1.05, 1),  # position sur X et Y
        loc='upper left',          # ancrage
        borderaxespad=0.
    )

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# ---------------------------
# Graphe pour RVM – BEL et APP
# ---------------------------
plot_noap_par_motif(df, "RVM", "BEL")
plot_noap_par_motif(df, "RVM", "APP")
```

```python
import matplotlib.pyplot as plt

def plot_delai_traitement(df, circuit, start_date, end_date):
    # Filtrer selon le circuit et la période
    subset = df[
        (df["Circuit"] == circuit) &
        (df["date_reception_vop"].dt.date >= start_date) &
        (df["date_reception_vop"].dt.date <= end_date)
    ]

    stats = subset.groupby(subset["date_reception_vop"].dt.date).agg(
        delai_moyen=("DelaiTraitement", "mean"),
        nb_requetes=("IdentifiantRequeteVOP", "count")
    ).reset_index()

    # Création du plot
    fig, ax1 = plt.subplots(figsize=(10,6))

    # Axe 1 : délai moyen (ligne bleue)
    ax1.plot(stats["date_reception_vop"], stats["delai_moyen"], marker="o", color="tab:blue", label="Délai moyen (sec)")
    ax1.set_xlabel("Date de réception VOP")
    ax1.set_ylabel("Délai moyen (sec)", color="tab:blue")
    ax1.tick_params(axis="y", labelcolor="tab:blue")

    # Axe 2 : nombre de requêtes (barres oranges)
    ax2 = ax1.twinx()
    ax2.bar(stats["date_reception_vop"], stats["nb_requetes"], alpha=0.3, color="tab:orange", label="Nb requêtes")
    ax2.set_ylabel("Nombre de requêtes", color="tab:orange")
    ax2.tick_params(axis="y", labelcolor="tab:orange")

    # Titre clair
    plt.title(f"Évolution du délai moyen de traitement - Circuit {circuit}", fontsize=14, weight="bold")

    # Légende hors du graphe
    lines, labels = ax1.get_legend_handles_labels()
    bars, bar_labels = ax2.get_legend_handles_labels()
    plt.legend(lines + bars, labels + bar_labels, loc="upper center", bbox_to_anchor=(0.5, -0.1), ncol=2)

    plt.tight_layout()
    plt.show()
```

```python
# Exemple sur 2 circuits
plot_delai_traitement(df, "RVM", start_date, end_date)
plot_delai_traitement(df, "On Us", start_date, end_date)
```
