````python
from datetime import date, timedelta
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------
# Fonction pour calculer et tracer % VOP LBP par statut
# ---------------------------
def plot_pct_vop_lbp(df, end_date, days=4):
    """
    Calcule et trace le pourcentage des réponses VOP émises vers LBP
    par statut et par date, limité au top 10 des confrères hors LBP.
    """
    # 1️⃣ Définir la période
    start_date = end_date - timedelta(days=days-1)

    # 2️⃣ Filtrer sur la période
    subset = df[
        (df["date_reception_vop"].dt.date >= start_date) &
        (df["date_reception_vop"].dt.date <= end_date)
    ]

    # 3️⃣ Séparer LBP et Hors LBP
    LBP_CODE = "PSSTFPPXXX"  # BIC La Banque Postale
    df_lbp = subset[subset["BicRonneurOrdre"] == LBP_CODE]
    df_hors_lbp = subset[subset["BicTeneurComate"] != LBP_CODE]

    # 4️⃣ Top 10 confrères hors LBP
    top10_confreres = (
        df_hors_lbp["BicTeneurComate"]
        .value_counts()
        .head(10)
        .index
    )
    df_hors_lbp = df_hors_lbp[df_hors_lbp["BicTeneurComate"].isin(top10_confreres)]

    # 5️⃣ Volume LBP par jour/statut
    vol_lbp = (
        df_lbp
        .groupby([df_lbp["date_reception_vop"].dt.date, "statut"])
        .size()
        .reset_index(name="count_lbp")
    )

    # 6️⃣ Volume hors LBP par jour/statut
    vol_hors_lbp = (
        df_hors_lbp
        .groupby([df_hors_lbp["date_reception_vop"].dt.date, "statut"])
        .size()
        .reset_index(name="count_hors_lbp")
    )

    # 7️⃣ Fusionner les deux (LBP + hors LBP)
    merged = pd.merge(
        vol_hors_lbp,
        vol_lbp,
        on=["date_reception_vop", "statut"],
        how="outer"
    ).fillna(0)

    merged["count_total"] = merged["count_lbp"] + merged["count_hors_lbp"]
    merged["pct_lbp"] = (merged["count_lbp"] / merged["count_total"]) * 100

    # 8️⃣ S’assurer que toutes les dates/statuts apparaissent (même si 0)
    all_combinations = pd.MultiIndex.from_product(
        [pd.date_range(start=start_date, end=end_date).date,
         ["match","no match","almost match","NOAP"]],  # tes 4 statuts
        names=["date_reception_vop", "statut"]
    )
    merged = merged.set_index(["date_reception_vop", "statut"]).reindex(all_combinations, fill_value=0).reset_index()

    # 9️⃣ Plot
    plt.figure(figsize=(10,6))
    sns.barplot(data=merged, x="date_reception_vop", y="pct_lbp", hue="statut")

    plt.title(f"% VOP LBP par statut ({days} jours)")
    plt.xlabel("Date de réception VOP")
    plt.ylabel("% VOP LBP")
    plt.xticks(rotation=45)

    plt.legend(
        title="Statut",
        bbox_to_anchor=(1.05, 1),
        loc='upper left',
        borderaxespad=0.
    )

    plt.tight_layout()
    plt.show()

    return merged  # renvoie le tableau pour l’exploiter si besoin
````

````python
import pandas as pd
import matplotlib.pyplot as plt

def plot_temps_moyen_par_circuit(df, circuit, start_date, end_date):
    # Filtrer par circuit et date
    subset = df[
        (df["Circuit"] == circuit) &
        (df["date_reception_vop"].dt.date >= start_date) &
        (df["date_reception_vop"].dt.date <= end_date)
    ].copy()

    # Vérifier les colonnes
    if "delai_traitement" not in subset.columns or "IdentifiantRequeteVOP" not in subset.columns:
        raise ValueError("Colonnes 'delai_traitement' ou 'IdentifiantRequeteVOP' manquantes !")

    # Grouper par date et calculer stats
    stats = subset.groupby(subset["date_reception_vop"].dt.date).agg({
        "delai_traitement": "mean",
        "IdentifiantRequeteVOP": "count"
    }).reset_index()

    # Renommer colonnes
    stats = stats.rename(columns={
        "delai_traitement": "temps_moyen",
        "IdentifiantRequeteVOP": "nb_requetes"
    })

    # Compléter tous les jours
    all_days = pd.date_range(start=start_date, end=end_date).date
    stats = stats.set_index("date_reception_vop").reindex(all_days, fill_value=0).rename_axis('date').reset_index()

    # Plot
    fig, ax1 = plt.subplots(figsize=(10,6))
    ax1.plot(stats["date"], stats["temps_moyen"], marker="o", color="tab:blue", label="Temps moyen (sec)")
    ax1.set_xlabel("Date")
    ax1.set_ylabel("Temps moyen (sec)", color="tab:blue")
    ax1.tick_params(axis="y", labelcolor="tab:blue")

    ax2 = ax1.twinx()
    ax2.bar(stats["date"], stats["nb_requetes"], alpha=0.3, color="tab:orange", label="Nb requêtes")
    ax2.set_ylabel("Nombre de requêtes", color="tab:orange")
    ax2.tick_params(axis="y", labelcolor="tab:orange")

    plt.title(f"Temps moyen et nb requêtes - Circuit {circuit}")
    plt.xticks(rotation=45)
    fig.autofmt_xdate()
    plt.tight_layout()
    plt.show()


# -------------------
# Exemple d'utilisation
# -------------------
start_date = pd.to_datetime("2025-09-01").date()
end_date = pd.to_datetime("2025-09-10").date()

# Graphe RVM
plot_temps_moyen_par_circuit(df, "RVM", start_date, end_date)

# Graphe On Us
plot_temps_moyen_par_circuit(df, "On Us", start_date, end_date)
````

````python
def plot_temps_moyen_par_jour_circuit(df, circuit):
    # Filtrer uniquement par circuit
    subset = df[df["Circuit"] == circuit]

    # Créer la liste de toutes les dates
    all_dates = [start_date + timedelta(days=i) for i in range((end_date - start_date).days + 1)]

    # Calculer le temps moyen pour chaque jour
    dates_plot = []
    temps_moyen_plot = []

    for d in all_dates:
        subset_jour = subset[subset["date_reception_vop"].dt.date == d]
        if not subset_jour.empty:
            moyenne = subset_jour["D-E-N-A-I-T-R-A-I-T-E-M-E-N-T"].mean()
        else:
            moyenne = 0  # pas de données pour ce jour
        dates_plot.append(d)
        temps_moyen_plot.append(moyenne)

    # ---------------------------
    # Plot
    # ---------------------------
    plt.figure(figsize=(10,6))
    plt.bar(dates_plot, temps_moyen_plot, color="skyblue")

    plt.title(f"Temps moyen de traitement par jour\nCircuit {circuit} (5 jours)")
    plt.xlabel("Date de réception")
    plt.ylabel("Temps moyen de traitement")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


# ---------------------------
# Graphe Temps moyen : Circuit RD1
# ---------------------------
plot_temps_moyen_par_jour_circuit(df, "RD1")
````

````python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import timedelta
import matplotlib.dates as mdates


def plot_temps_moyen_par_jour(df, start_date, end_date):
    """
    Trace le temps moyen de traitement par jour pour chaque circuit
    dans la période donnée.

    df : DataFrame avec au moins ['circuit', 'date_reception_vop', 'delai_traitement']
    start_date, end_date : objets datetime.date
    """

    # Créer toutes les dates de la période
    all_dates = [start_date + timedelta(days=i) for i in range((end_date - start_date).days + 1)]

    circuits = df["circuit"].unique()

    # Préparer les données pour le graphique
    plot_data = {}

    for circuit in circuits:
        subset = df[df["circuit"] == circuit]
        temps_moyen_plot = []

        for d in all_dates:
            subset_jour = subset[subset["date_reception_vop"].dt.date == d]
            if not subset_jour.empty:
                moyenne = subset_jour["delai_traitement"].mean()
            else:
                moyenne = 0
            temps_moyen_plot.append(moyenne)

        plot_data[circuit] = temps_moyen_plot

    # Plot
    plt.figure(figsize=(12,6))

    for circuit, valeurs in plot_data.items():
        plt.plot(all_dates, valeurs, marker='o', label=f'Circuit {circuit}')

    plt.title("Temps moyen de traitement par jour et par circuit")
    plt.xlabel("Date de réception")
    plt.ylabel("Temps moyen de traitement")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
````

````python


def calculer_temps_moyen_par_jour(df, start_date, end_date):
    """
    Calcule le temps moyen de traitement et le nombre de requêtes par jour et par circuit.

    df : DataFrame avec au moins ['circuit', 'date_reception_vop', 'delai_traitement']
    start_date, end_date : objets datetime.date
    """
    all_dates = [start_date + timedelta(days=i) for i in range((end_date - start_date).days + 1)]
    circuits = df["circuit"].unique()

    result = []
    for circuit in circuits:
        subset = df[df["circuit"] == circuit]
        for d in all_dates:
            subset_jour = subset[subset["date_reception_vop"].dt.date == d]
            if not subset_jour.empty:
                moyenne = subset_jour["delai_traitement"].mean()
                nb_requetes = subset_jour.shape[0]
            else:
                moyenne = 0
                nb_requetes = 0
            result.append({
                "date": d,
                "circuit": circuit,
                "temps_moyen": moyenne,
                "nb_requetes": nb_requetes
            })

    return pd.DataFrame(result)

def plot_temps_moyen(stats):
    circuits = stats["circuit"].unique()
    plt.figure(figsize=(12,6))

    for circuit in circuits:
        subset = stats[stats["circuit"] == circuit]
        # Convertir en datetime avec heure à 00:00 pour éviter l'affichage du temps
        dates = pd.to_datetime(subset["date"])
        plt.plot(dates, subset["temps_moyen"], marker='o', label=f'Circuit {circuit}')

    # Formatter l'axe X : jour, mois, année
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator())

    plt.title("Temps moyen de traitement par jour et par circuit")
    plt.xlabel("Date de réception")
    plt.ylabel("Temps moyen de traitement")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
import pandas as pd
from datetime import datetime, date

# Création d'un petit jeu de données
data = {
    "circuit": ["A", "A", "A", "B", "B", "C"],
    "date_reception_vop": [
        datetime(2025, 1, 1, 10),
        datetime(2025, 1, 1, 15),
        datetime(2025, 1, 3, 11),
        datetime(2025, 1, 2, 9),
        datetime(2025, 1, 3, 14),
        datetime(2025, 1, 5, 16)
    ],
    "delai_traitement": [5, 7, 6, 8, 4, 10]
}

df = pd.DataFrame(data)

# Vérifier le DataFrame
print(df)

start_date = date(2025, 1, 1)
end_date = date(2025, 1, 5)

# Calculer les temps moyens
stats = calculer_temps_moyen_par_jour(df, start_date, end_date)
print(stats)

# Tracer le graphique
plot_temps_moyen(stats)

stats = calculer_temps_moyen_par_jour(df, start_date, end_date)

# 3️⃣ Tracer le graphique
plot_temps_moyen(stats)
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# 1. Fonction : calcul des pourcentages
# ----------------------------
def calculer_pourcentages(df, bic_art="PSSTFRPPXXX", date_ref=None, nb_jours=4):
    """
    Calcule le pourcentage d'équipements COV par statut et par confrère,
    uniquement pour bic_teneur_art = bic_art et en excluant LBP côté confrères.
    On peut filtrer sur une fenêtre de dates autour d'une date de référence.
    """
    # Filtrer sur notre banque
    df_filtre = df[df["bic_teneur_art"] == bic_art]
    # Exclure LBP côté confrères
    df_filtre = df_filtre[df_filtre["bic_teneur_compte"] != bic_art]

    # Filtrer sur la fenêtre de dates si date_ref est définie
    if date_ref is not None:
        date_ref = pd.to_datetime(date_ref)
        date_min = date_ref - pd.Timedelta(days=nb_jours)
        date_max = date_ref
        df_filtre = df_filtre[(df_filtre["date"] >= date_min) & (df_filtre["date"] <= date_max)]

    # Compter le nombre de cas par date + confrère + statut
    df_grouped = df_filtre.groupby(["date", "bic_teneur_compte", "statut"]).size().reset_index(name="nb")

    # Ajouter le total par date + confrère
    df_grouped["total"] = df_grouped.groupby(["date", "bic_teneur_compte"])["nb"].transform("sum")

    # Calcul du pourcentage
    df_grouped["pourcentage"] = (df_grouped["nb"] / df_grouped["total"]) * 100

    return df_grouped

# ----------------------------
# 2. Fonction : sélectionner top 10 confrères
# ----------------------------
def top10_confreres(df):
    """
    Retourne uniquement les 10 confrères avec le plus de volume total.
    """
    top10 = df.groupby("bic_teneur_compte")["nb"].sum().nlargest(10).index
    return df[df["bic_teneur_compte"].isin(top10)]

# ----------------------------
# 3. Fonction : plot
# ----------------------------
def plot_pourcentages(df, titre="Répartition des statuts"):
    """
    Affiche un plot empilé des pourcentages par statut et par jour.
    """
    pivot = df.groupby(["date", "statut"])["nb"].sum().reset_index()
    pivot_total = pivot.groupby("date")["nb"].transform("sum")
    pivot["pourcentage"] = pivot["nb"] / pivot_total * 100

    # Transformer pour plot
    pivot_plot = pivot.pivot(index="date", columns="statut", values="pourcentage").fillna(0)
    pivot_plot = pivot_plot.sort_index()  # tri par date

    pivot_plot.plot(kind="bar", stacked=True, figsize=(10, 6))
    plt.title(titre)
    plt.ylabel("Pourcentage (%)")
    plt.legend(title="Statut", bbox_to_anchor=(1.05, 1))
    plt.tight_layout()
    plt.show()

# ----------------------------
# 4. Exemple d'utilisation
# ----------------------------
# Dataset jouet
data = {
    "date": pd.to_datetime(
        ["2025-06-20","2025-06-20","2025-06-20","2025-06-21","2025-06-21",
         "2025-06-21","2025-06-22","2025-06-22","2025-06-23","2025-06-23"]
    ),
    "bic_teneur_art": ["PSSTFRPPXXX"]*10,
    "bic_teneur_compte": ["BNPAFRPPXXX","BNPAFRPPXXX","SOGEFRPPXXX","BNPAFRPPXXX",
                          "LCLMFRPPXXX","AGRIFRPPXXX","BNPAFRPPXXX","SOGEFRPPXXX",
                          "CMBRFR2BXXX","HSBCFRPPXXX"],
    "statut": ["NMTH","MTCH","NMTH","CMTCCH","NMTH","NOAP","MTCH","NMTH","CMTCCH","NOAP"]
}
df = pd.DataFrame(data)

# Définir la date de référence
date_ref = "2025-06-23"

# Calcul des pourcentages sur 4 jours avant date_ref
df_pct = calculer_pourcentages(df, date_ref=date_ref, nb_jours=4)
print("Table complète filtrée sur 4 jours avant la date de référence :")
print(df_pct)

# Top 10 confrères
df_top10 = top10_confreres(df_pct)
print("\nTop 10 confrères :")
print(df_top10)

# Plot
plot_pourcentages(df_top10, titre="Répartition des statuts - Top 10 confrères (4 jours)")
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# 1. Fonction : calcul du temps moyen par circuit et par jour (multi-circuits)
# ----------------------------
def calcul_temps_moyen_multi(df, circuits=["RVM", "ONUS"], date_ref=None, nb_jours=4):
    """
    Calcule le temps moyen de traitement par circuit et par date pour plusieurs circuits.
    - df : dataframe contenant ['date', 'circuit', 'delai_rvm']
    - circuits : liste des circuits à filtrer (ex: ['RVM', 'ONUS'])
    - date_ref : date de référence pour ne prendre que les nb_jours précédents
    """
    # Filtrer sur les circuits
    df_filtre = df[df["circuit"].isin(circuits)]

    # Filtrer sur la fenêtre de dates si date_ref est défini
    if date_ref is not None:
        date_ref = pd.to_datetime(date_ref)
        date_min = date_ref - pd.Timedelta(days=nb_jours)
        date_max = date_ref
        df_filtre = df_filtre[(df_filtre["date"] >= date_min) & (df_filtre["date"] <= date_max)]

    # Calcul du temps moyen et du nombre de lignes par date et circuit
 stats = (
    df_filtre.groupby(["date", "circuit"])
    .apply(lambda g: pd.Series({
        "temps_moyen": g["delai_rvm"].mean(),
        "nb_lignes": g["delai_rvm"].count()
    }))
    .reset_index()
    .rename(columns={"circuit": "Circuit"})
)

    return stats

# ----------------------------
# 2. Fonction : plot multi-circuits
# ----------------------------
def plot_temps_moyen(stats):
    plt.figure(figsize=(8,5))

    for circuit in stats["Circuit"].unique():
        subset = stats[stats["Circuit"] == circuit]
        plt.plot(subset["date"], subset["temps_moyen"], marker="o", label=circuit)

    plt.title("Temps moyen de traitement par circuit")
    plt.xlabel("Date")
    plt.ylabel("Temps moyen de traitement")
    plt.legend(title="Circuit")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()

# ----------------------------
# 3. Exemple d'utilisation
# ----------------------------
data = {
    "date": pd.to_datetime([
        "2025-06-20","2025-06-20","2025-06-21","2025-06-21",
        "2025-06-22","2025-06-22","2025-06-23","2025-06-23",
        "2025-06-20","2025-06-21","2025-06-22","2025-06-23"
    ]),
    "circuit": ["RVM","RVM","RVM","RVM","RVM","RVM","RVM","RVM",
                "ONUS","ONUS","ONUS","ONUS"],
    "delai_rvm": [12, 15, 11, 14, 13, 16, 10, 12, 20, 18, 19, 21]
}
df = pd.DataFrame(data)

# Date de référence
date_ref = "2025-06-23"

# Calcul sur 4 jours avant la date de référence
stats = calcul_temps_moyen_multi(df, circuits=["RVM","ONUS"], date_ref=date_ref, nb_jours=4)
print(stats)

# Plot
plot_temps_moyen(stats)
````

````python
import pandas as pd

def taux_vop_impossible_emission_par_jour(df: pd.DataFrame, end_date: str | pd.Timestamp, nb_jours: int = 4) -> pd.DataFrame:
    """
    Calcule le taux de VOP impossible (NOAP + F004) en émission par jour
    sur une fenêtre [end_date - nb_jours, end_date].
    La colonne date est supposée déjà en datetime (UNDATE).
    """
    # Conversion de end_date si besoin
    end_date = pd.to_datetime(end_date)
    start_date = end_date - pd.Timedelta(days=nb_jours)

    # Filtrer période + périmètre émission
    df_filtre = df[
        (df["UNDATE"] >= start_date) &
        (df["UNDATE"] <= end_date) &
        (df["bic_donneur"] == "BTSSTFRPPXXX") &
        (df["bic_teneur"] != "PSSTFRPPXXX")
    ].copy()

    if df_filtre.empty:
        return pd.DataFrame(columns=["UNDATE", "total_reponses", "nb_vop_impossible", "taux_%"])

    # Grouper par jour
    stats = (
        df_filtre.groupby("UNDATE")
        .apply(lambda g: pd.Series({
            "total_reponses": len(g),
            "nb_vop_impossible": len(g[(g["statut"] == "NOAP") & (g["motif"] == "F004")])
        }))
        .reset_index()
    )

    # Calcul du taux
    stats["taux_%"] = stats["nb_vop_impossible"] / stats["total_reponses"] * 100
    return stats
````

````python
import pandas as pd
import matplotlib.pyplot as plt

def taux_vop_possible_emission_par_jour(df: pd.DataFrame, date_ref=None, nb_jours: int = 4) -> pd.DataFrame:
    """
    Calcule le taux de VOP possible en émission par jour
    sur une fenêtre [date_ref - nb_jours, date_ref].
    Colonnes attendues : date_, bic_6_donneur_ordre, bic_8_teneur_ordre, statut, motif.
    """
    # 1. Gestion des dates
    if date_ref is not None:
        date_ref = pd.to_datetime(date_ref)
        date_min = date_ref - pd.Timedelta(days=nb_jours)
        date_max = date_ref
        df = df[(df["date_"] >= date_min) & (df["date_"] <= date_max)]

    # 2. Filtre émission : donneur = LBP, teneur ≠ LBP
    df_filtre = df[
        (df["bic_6_donneur_ordre"] == "BTSSTFRPPXXX") &
        (df["bic_8_teneur_ordre"] != "PSSTFRPPXXX")
    ].copy()

    if df_filtre.empty:
        return pd.DataFrame(columns=["date_", "total_reponses", "nb_vop_possible", "taux_%"])

    # 3. Grouper par date_ et calculer
    stats = (
        df_filtre.groupby("date_")
        .apply(lambda g: pd.Series({
            "total_reponses": len(g),
            "nb_vop_possible": len(g[(g["statut"] != "NOAP") | (g["motif"] != "F004")])
        }))
        .reset_index()
    )

    # 4. Ajout du taux
    stats["taux_%"] = stats["nb_vop_possible"] / stats["total_reponses"] * 100
    stats = stats.sort_values("date_")

    return stats

def plot_taux_vop(df_stats: pd.DataFrame, title: str = "Taux de VOP possible en émission par jour"):
    """
    Affiche un plot du taux de VOP possible en émission par jour.
    X = date_, Y = taux_%
    """
    if df_stats.empty:
        print("Aucune donnée à afficher.")
        return

    plt.figure(figsize=(10, 5))
    plt.plot(df_stats["date_"], df_stats["taux_%"], marker='o', linestyle='-')
    plt.xticks(rotation=45)
    plt.xlabel("Date")
    plt.ylabel("Taux (%)")
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
````
