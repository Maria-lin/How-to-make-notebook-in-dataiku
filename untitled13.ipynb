```python
SELECT
  CAST(SUBSTR(date_operation, 1, 2) AS INT) AS jour,
  CASE
    WHEN SUBSTR(num_carte, 1, 4) = '4970' THEN 'Visa Realys ou Classic 4970'
    WHEN SUBSTR(num_carte, 1, 4) = '4973' THEN 'Visa Premier 4973'
    WHEN SUBSTR(num_carte, 1, 4) = '4974' THEN 'Visa Infinite 4974'
    ELSE 'Non-Porteur'
  END AS type_porteur,
  SUM(CAST(montant_ope AS DOUBLE)) / 100 AS montant_total
FROM trepor_an0.t_dlk_mon_nf2_ope_retrait
WHERE
  CAST(SUBSTR(date_operation, 5, 4) AS INT) = 2025
  AND CAST(SUBSTR(date_operation, 3, 2) AS INT) = 12
GROUP BY
  CAST(SUBSTR(date_operation, 1, 2) AS INT),
  CASE
    WHEN SUBSTR(num_carte, 1, 4) = '4970' THEN 'Visa Realys ou Classic 4970'
    WHEN SUBSTR(num_carte, 1, 4) = '4973' THEN 'Visa Premier 4973'
    WHEN SUBSTR(num_carte, 1, 4) = '4974' THEN 'Visa Infinite 4974'
    ELSE 'Non-Porteur'
  END
ORDER BY
  jour,
  type_porteur;
```

```python
# ============================
# üì¶ Imports
# ============================
import pandas as pd
import matplotlib.pyplot as plt

# ============================
# üß™ 1) Dataset fictif (exemple)
# ============================
data = {
    "num_automate": [
        "GAB001","GAB002","GAB003","GAB004","GAB005",
        "GAB006","GAB007","GAB008","GAB009","GAB010"
    ],
    "code_postale_emplacement": [
        "75001","75001","75002","75002","75002",
        "13001","13001","33000","33000","33000"
    ],
    "ville": [
        "Paris","Paris","Paris","Paris","Paris",
        "Marseille","Marseille","Bordeaux","Bordeaux","Bordeaux"
    ]
}

df_rea_gab = pd.DataFrame(data)

# ============================
# üßÆ 2) Fonction ‚Äî compter GAB par code postal
# ============================
def compter_gab_par_code_postal(df: pd.DataFrame) -> pd.DataFrame:
    res = (
        df.groupby("code_postale_emplacement")["num_automate"]
          .nunique()
          .reset_index(name="nb_gab")
          .sort_values("nb_gab", ascending=False)
          .reset_index(drop=True)
    )
    return res

gab_par_cp = compter_gab_par_code_postal(df_rea_gab)

# ============================
# üìä 3) Fonction ‚Äî plot Top N
# ============================
def plot_top_code_postal_gab(gab_par_cp, top_n=20):
    top = gab_par_cp.head(top_n)

    plt.figure(figsize=(10, 5))
    plt.bar(top["code_postale_emplacement"].astype(str), top["nb_gab"])
    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Code postal emplacement")
    plt.ylabel("Nombre de GAB")
    plt.title(f"Top {top_n} codes postaux par nombre de GAB")
    plt.tight_layout()
    plt.show()

# ============================
# ‚ñ∂Ô∏è 4) Ex√©cution
# ============================
print(gab_par_cp)
plot_top_code_postal_gab(gab_par_cp, top_n=10)
```
