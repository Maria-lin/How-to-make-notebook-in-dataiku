```python
SELECT
  CAST(SUBSTR(date_operation, 1, 2) AS INT) AS jour,
  CASE
    WHEN SUBSTR(num_carte, 1, 4) = '4970' THEN 'Visa Realys ou Classic 4970'
    WHEN SUBSTR(num_carte, 1, 4) = '4973' THEN 'Visa Premier 4973'
    WHEN SUBSTR(num_carte, 1, 4) = '4974' THEN 'Visa Infinite 4974'
    ELSE 'Non-Porteur'
  END AS type_porteur,
  SUM(CAST(montant_ope AS DOUBLE)) / 100 AS montant_total
FROM trepor_an0.t_dlk_mon_nf2_ope_retrait
WHERE
  CAST(SUBSTR(date_operation, 5, 4) AS INT) = 2025
  AND CAST(SUBSTR(date_operation, 3, 2) AS INT) = 12
GROUP BY
  CAST(SUBSTR(date_operation, 1, 2) AS INT),
  CASE
    WHEN SUBSTR(num_carte, 1, 4) = '4970' THEN 'Visa Realys ou Classic 4970'
    WHEN SUBSTR(num_carte, 1, 4) = '4973' THEN 'Visa Premier 4973'
    WHEN SUBSTR(num_carte, 1, 4) = '4974' THEN 'Visa Infinite 4974'
    ELSE 'Non-Porteur'
  END
ORDER BY
  jour,
  type_porteur;
```

```python
# ============================================================
# üìä REPORTING MODERNE : R√©partition des GAB par code postal
# Compatible Python 3.6 (DSS, kernel python36)
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# ------------------------------------------------------------
# 0) Chargement des donn√©es
#    üëâ Si df_rea_gab existe d√©j√†, on l'utilise.
#    üëâ Sinon, on cr√©e un dataset fictif pour l'exemple.
# ------------------------------------------------------------
try:
    df_rea_gab
    data_source_msg = "‚úÖ Utilisation du DataFrame existant : df_rea_gab"
except NameError:
    data_source_msg = "‚ÑπÔ∏è df_rea_gab n'existe pas encore, cr√©ation d'un dataset fictif pour l'exemple."
    data = {
        "num_automate": [
            "GAB001", "GAB002", "GAB003", "GAB004", "GAB005",
            "GAB006", "GAB007", "GAB008", "GAB009", "GAB010"
        ],
        "code_postale_emplacement": [
            "75001", "75001", "75002", "75002", "75002",
            "13001", "13001", "33000", "33000", "33000"
        ],
        "ville": [
            "Paris", "Paris", "Paris", "Paris", "Paris",
            "Marseille", "Marseille", "Bordeaux", "Bordeaux", "Bordeaux"
        ]
    }
    df_rea_gab = pd.DataFrame(data)

# Style global des graphiques (un peu plus moderne)
plt.style.use("ggplot")

# ------------------------------------------------------------
# 1) Fonctions utilitaires
# ------------------------------------------------------------
def compter_gab_par_code_postal(df):
    """
    Retourne un DataFrame avec le nombre de GAB uniques par code postal.
    """
    res = (
        df.groupby("code_postale_emplacement")["num_automate"]
          .nunique()
          .reset_index(name="nb_gab")
          .sort_values("nb_gab", ascending=False)
          .reset_index(drop=True)
    )
    return res


def afficher_kpis_gab(df, gab_par_cp):
    """
    Affiche des KPI propres sous forme de Markdown.
    """
    nb_gab_total = df["num_automate"].nunique()
    nb_cp = gab_par_cp["code_postale_emplacement"].nunique()
    moyenne_par_cp = gab_par_cp["nb_gab"].mean()

    display(Markdown("### üí° Indicateurs cl√©s"))
    display(Markdown(
        "- **Nombre total de GAB uniques :** `{}`\n"
        "- **Nombre de codes postaux couverts :** `{}`\n"
        "- **Nombre moyen de GAB par code postal :** `{:.2f}`"
        .format(nb_gab_total, nb_cp, moyenne_par_cp)
    ))


def plot_top_code_postal_gab(gab_par_cp, top_n=20, titre=None):
    """
    Affiche un graphique moderne (barres horizontales) du top N codes postaux.
    """
    top = gab_par_cp.head(top_n).copy()
    top = top.sort_values("nb_gab", ascending=True)  # pour un barh lisible

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.barh(top["code_postale_emplacement"].astype(str), top["nb_gab"])
    ax.set_xlabel("Nombre de GAB")
    ax.set_ylabel("Code postal emplacement")

    if titre is None:
        titre = "Top {} codes postaux par nombre de GAB".format(top_n)
    ax.set_title(titre)

    # Ajout des valeurs au bout des barres
    for i, (cp, nb) in enumerate(zip(top["code_postale_emplacement"], top["nb_gab"])):
        ax.text(nb + 0.05, i, str(nb), va="center")

    plt.tight_layout()
    plt.show()


def afficher_table_styled(gab_par_cp, top_n=20):
    """
    Affiche un tableau styl√© (pandas Styler) pour les top N codes postaux.
    """
    top = gab_par_cp.head(top_n).copy()
    styled = (
        top.style
           .background_gradient(subset=["nb_gab"], cmap="Blues")
           .set_table_styles(
               [
                   {"selector": "th", "props": [("background-color", "#1f2933"), ("color", "white")]},
                   {"selector": "td", "props": [("border", "1px solid #ddd")]}
               ]
           )
           .set_properties(**{"text-align": "center"})
    )
    display(Markdown("### üìã Top {} codes postaux par nombre de GAB".format(top_n)))
    display(styled)


# ------------------------------------------------------------
# 2) Pipeline complet de reporting
# ------------------------------------------------------------

# Titre principal du reporting
display(Markdown("# üìä Reporting GAB par code postal"))
display(Markdown("_{}_".format(data_source_msg)))

# Calcul des agr√©gations
gab_par_cp = compter_gab_par_code_postal(df_rea_gab)

# Affichage des KPI
afficher_kpis_gab(df_rea_gab, gab_par_cp)

# Tableau styl√©
afficher_table_styled(gab_par_cp, top_n=20)

# Graphique
plot_top_code_postal_gab(
    gab_par_cp,
    top_n=20,
    titre="Top 20 codes postaux par nombre de GAB"
)
```

```python
WITH base AS (
  SELECT
    circuit,
    -- Parse ISO 8601: 2026-01-04T00:01:39.551Z  -> timestamp
    from_unixtime(
      unix_timestamp(`Timestamp Reception VOP`, "yyyy-MM-dd'T'HH:mm:ss.SSS'Z'")
    ) AS vop_ts
  FROM ma_table
),
filtered AS (
  SELECT
    circuit,
    vop_ts
  FROM base
  WHERE vop_ts >= CAST('2025-10-09 00:00:00' AS TIMESTAMP)
)
SELECT
  circuit,
  -- d√©but du mois (DATE): 2025-10-01, 2025-11-01, ...
  trunc(to_date(vop_ts), 'MM') AS mois,
  COUNT(1) AS nb_requetes_vop
FROM filtered
GROUP BY
  circuit,
  trunc(to_date(vop_ts), 'MM')
ORDER BY
  mois,
  circuit;
```

```python
SELECT
  annee,
  rubrique,
  volume,
  volume_cpp,
  volume_epa
FROM
(
  /* 1) CPP/CCP ‚Äì GAB internes */
  SELECT
    CAST(substr(DateOperation, 1, 4) AS INT) AS annee,
    'Nombre de retraits Cartes bancaires Post (CPP/CCP) ‚Äì GAB internes' AS rubrique,
    SUM(CASE WHEN CodeCartePrivative NOT IN ('EPA','COS') THEN 1 ELSE 0 END) AS volume,
    SUM(CASE WHEN CodeCartePrivative NOT IN ('EPA','COS') THEN 1 ELSE 0 END) AS volume_cpp,
    0 AS volume_epa
  FROM votre_table
  WHERE substr(DateOperation, 1, 4) IN ('2023','2024')
    AND (GAB_interne = 1 OR upper(GAB_interne) = 'OUI' OR upper(GAB_interne) = 'TRUE')
  GROUP BY CAST(substr(DateOperation, 1, 4) AS INT)

  UNION ALL

  /* 2) CPP/CCP ‚Äì Autres GAB */
  SELECT
    CAST(substr(DateOperation, 1, 4) AS INT) AS annee,
    'Nombre de retraits Cartes bancaires Post (CPP/CCP) ‚Äì Autres GAB' AS rubrique,
    SUM(CASE WHEN CodeCartePrivative NOT IN ('EPA','COS') THEN 1 ELSE 0 END) AS volume,
    SUM(CASE WHEN CodeCartePrivative NOT IN ('EPA','COS') THEN 1 ELSE 0 END) AS volume_cpp,
    0 AS volume_epa
  FROM votre_table
  WHERE substr(DateOperation, 1, 4) IN ('2023','2024')
    AND NOT (GAB_interne = 1 OR upper(GAB_interne) = 'OUI' OR upper(GAB_interne) = 'TRUE')
  GROUP BY CAST(substr(DateOperation, 1, 4) AS INT)

  UNION ALL

  /* 3) PostEpargne ‚Äì GAB internes */
  SELECT
    CAST(substr(DateOperation, 1, 4) AS INT) AS annee,
    'Nombre de retraits PostEpargne (EPA) ‚Äì GAB internes' AS rubrique,
    SUM(CASE WHEN CodeCartePrivative = 'EPA' THEN 1 ELSE 0 END) AS volume,
    0 AS volume_cpp,
    SUM(CASE WHEN CodeCartePrivative = 'EPA' THEN 1 ELSE 0 END) AS volume_epa
  FROM votre_table
  WHERE substr(DateOperation, 1, 4) IN ('2023','2024')
    AND (GAB_interne = 1 OR upper(GAB_interne) = 'OUI' OR upper(GAB_interne) = 'TRUE')
  GROUP BY CAST(substr(DateOperation, 1, 4) AS INT)

  UNION ALL

  /* 4) PostEpargne ‚Äì Autres GAB */
  SELECT
    CAST(substr(DateOperation, 1, 4) AS INT) AS annee,
    'Nombre de retraits PostEpargne (EPA) ‚Äì Autres GAB' AS rubrique,
    SUM(CASE WHEN CodeCartePrivative = 'EPA' THEN 1 ELSE 0 END) AS volume,
    0 AS volume_cpp,
    SUM(CASE WHEN CodeCartePrivative = 'EPA' THEN 1 ELSE 0 END) AS volume_epa
  FROM votre_table
  WHERE substr(DateOperation, 1, 4) IN ('2023','2024')
    AND NOT (GAB_interne = 1 OR upper(GAB_interne) = 'OUI' OR upper(GAB_interne) = 'TRUE')
  GROUP BY CAST(substr(DateOperation, 1, 4) AS INT)
) t
ORDER BY annee, rubrique;
```
